```python
df["Review Counts"] = pd.to_numeric(df["Review Counts"], errors="coerce")
```

```python
bins = [
    0, 50, 100, 150, 200, 250, 300, 350, 400, 450, 500,
    600, 700, 800, 1000, 1500, 2000, float("inf")
]
labels = [
    "less_50",
    "51-100",
    "101-150",
    "151-200",
    "201-250",
    "251-300",
    "301-350",
    "351-400",
    "401-450",
    "451-500",
    "501-600",
    "601-700",
    "701-800",
    "801-1000",
    "1001-1500",
    "1501-2000",
    "2000_plus"
]
df["Review_Group"] = pd.cut(df["Review Counts"], bins=bins, labels=labels, right=True, include_lowest=True)
```

```python
df.head()
```

```python
df['Review_Group'].value_counts().sort_index()
```

```python
less_200_df = df[df['Review Counts']<= 200]
less_200_df = less_200_df[["Destination", "Links"]]
```
