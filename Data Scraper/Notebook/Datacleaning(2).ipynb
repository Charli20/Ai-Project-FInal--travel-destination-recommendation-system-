```python
import pandas as pd
import os
```

```python

folder_path = r"C:\\NIBM\\Travel recommendation system\\Travel-Recommendation-system-in-Sri-Lanka-\\dataset\\scaped_data"

output_csv = os.path.join(folder_path, "rating_distributions_summary.csv")


csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]


summary_data = []

for file in csv_files:
    file_path = os.path.join(folder_path, file)
    df = pd.read_csv(file_path)

    if 'Review Score' in df.columns: 
        total_reviews = len(df)
        rating_counts = df['Review Score'].value_counts().sort_index()  
        rating_percentages = (rating_counts / total_reviews * 100).round(2) 

        row = {
            "Destination Title": file.replace(".csv", ""),
            "Total Reviews (Destination)": total_reviews,
            "1 Star %": rating_percentages.get(1, 0.0),
            "2 Star %": rating_percentages.get(2, 0.0),
            "3 Star %": rating_percentages.get(3, 0.0),
            "4 Star %": rating_percentages.get(4, 0.0),
            "5 Star %": rating_percentages.get(5, 0.0),
        }
        summary_data.append(row)
    else:
        print(f"\nFile {file} does not have 'Review Score' column.")


summary_df = pd.DataFrame(summary_data)


overall_avg = summary_df[["1 Star %", "2 Star %", "3 Star %", "4 Star %", "5 Star %"]].mean().round(2)
overall_row = {
    "Destination Title": "Overall Average",
    "Total Reviews (Destination)": summary_df["Total Reviews (Destination)"].sum(),
    "1 Star %": overall_avg["1 Star %"],
    "2 Star %": overall_avg["2 Star %"],
    "3 Star %": overall_avg["3 Star %"],
    "4 Star %": overall_avg["4 Star %"],
    "5 Star %": overall_avg["5 Star %"],
}
summary_df = pd.concat([summary_df, pd.DataFrame([overall_row])], ignore_index=True)

summary_df.to_csv(output_csv, index=False, encoding="utf-8-sig")

print(f"\n✅ Rating distribution summary saved to: {output_csv}")

    
```

```python
import os
import pandas as pd
import numpy as np
import math
```

```python
# input_folder = r"C:\NIBM\Travel recommendation system\Travel-Recommendation-system-in-Sri-Lanka-\dataset\updated_csvs"
input_folder = r"C:\\NIBM\\Travel recommendation system\\Travel-Recommendation-system-in-Sri-Lanka-\\Data Preparation\\dataset\\updated_csvs"

output_folder = r"C:\\NIBM\\Travel recommendation system\\Travel-Recommendation-system-in-Sri-Lanka-\\Data Preparation\\dataset\\final_data"

max_reviews = 250
min_keep_reviews = 10  
```

```python
os.makedirs(output_folder, exist_ok=True)

for file in os.listdir(input_folder):
    if not file.endswith(".csv"):
        continue
    df = pd.read_csv(os.path.join(input_folder, file))
    
    total_reviews = df['Total Reviews (Destination)'].iloc[0]

    if total_reviews <= max_reviews:
        df.to_csv(os.path.join(output_folder, file), index=False)
        print(f"Saved {file} as-is, total reviews = {total_reviews}")
        continue

   
    downsampled_list = []

    # Count reviews per star
    star_counts = df['Review Score'].value_counts().to_dict()


    small_stars = {star: count for star, count in star_counts.items() if count < min_keep_reviews}
    for star, count in small_stars.items():
        downsampled_list.append(df[df['Review Score'] == star])

    # Remaining stars
    remaining_stars = [star for star in range(1, 6) if star not in small_stars]
    remaining_slots = max_reviews - sum(small_stars.values())
    remaining_total_reviews = sum(star_counts.get(star, 0) for star in remaining_stars)

 
    for star in remaining_stars:
        current_count = star_counts.get(star, 0)
        if current_count == 0:
            continue
      
        target_count = int(current_count / remaining_total_reviews * remaining_slots)
     
        target_count = min(target_count, current_count)
        sampled = df[df['Review Score'] == star].sample(n=target_count, random_state=42)
        downsampled_list.append(sampled)

    downsampled_df = pd.concat(downsampled_list).sample(frac=1, random_state=42).reset_index(drop=True)
    downsampled_df.to_csv(os.path.join(output_folder, file), index=False)
    
    print(f"Processed {file}: downsampled to {len(downsampled_df)} reviews")
```

```python
categories = ["Family", "Couples", "Solo", "Business", "Friends", "Missing"]

for fl in os.listdir(input_folder):
    if not fl.endswith(".csv"):
        continue

    file_path = os.path.join(input_folder, fl)
    df = pd.read_csv(file_path)


    df['Reviewer Category'] = df['Reviewer Category'].fillna("Missing")

  
    category_counts = df['Reviewer Category'].value_counts().to_dict()


    for c in categories:
        if c not in category_counts:
            category_counts[c] = 0

    total_reviews = len(df)


    category_percentages = {
        f"{c} %": (category_counts[c] / total_reviews) * 100
        for c in categories
    }

 
    percent_cols = [col for col in df.columns if col.endswith("%")]
    df.drop(columns=percent_cols, inplace=True, errors="ignore")

    for col_name, perc in category_percentages.items():
        df[col_name] = perc


    df.to_csv(file_path, index=False)
    print(f"Updated {fl} with reviewer category percentages (including Missing)")
```
