```python

df1_titles = df1['Destination Title'].astype(str).str.strip()
df2_titles = df2['Destination'].astype(str).str.strip()

missing_destinations = df2_titles[~df2_titles.isin(df1_titles)]


missing_destinations = missing_destinations.drop_duplicates()


```

```python
missing_df = pd.DataFrame(
    missing_destinations.tolist(),
    columns=["Missing Destinations"]
)

missing_df.to_csv(
    "missing_destinations.csv",
    index=False,
    encoding="utf-8"
)

print("CSV saved successfully!")
```
