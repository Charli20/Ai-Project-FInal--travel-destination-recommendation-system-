```python
import re
def text_preprocessing(text):
    text = text.lower()
    text = re.sub(r"<.*?>", " ", text)
    text = re.sub(r"[^\w\s]", " ", text)
    text = re.sub(r"\s+", " ", text).strip()

    return text
```

```python
df['clean_embedding'] = df['text_embedding'].apply(text_preprocessing)
```

Model Building

```python
from sentence_transformers import SentenceTransformer, util

# Initialize SBERT model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Encode your document texts (already cleaned)
doc_embeddings = model.encode(df["clean_embedding"].tolist(), convert_to_tensor=True)
print("Document embeddings shape:", doc_embeddings.shape)
```

```python
hidden_gem_threshold = 50  # max reviews to consider hidden gem
top_n_recommendations = 10

# Filter hidden gems
hidden_df = df[df['Total Reviews'] <= hidden_gem_threshold].copy()

query = "I want a beautiful mountain to climb in Sri Lanka"
query_emb = model.encode(query, convert_to_tensor=True)

# Compute cosine similarity for hidden gems
hidden_embeddings = model.encode(hidden_df["clean_embedding"].tolist(), convert_to_tensor=True)
similarities = util.cos_sim(query_emb, hidden_embeddings).cpu().numpy()[0]

hidden_df['similarity_score'] = similarities
hidden_df = hidden_df.sort_values(by='similarity_score', ascending=False).head(top_n_recommendations)

print("üåü Hidden Gems:")
hidden_df[['Destination Title', 'Total Reviews', 'similarity_score']]
```

```python
# Compute local similarities for all documents
similarities = util.cos_sim(query_emb, doc_embeddings).cpu().numpy()[0]
df['similarity_score'] = similarities

local_rec = df.sort_values(by='similarity_score', ascending=False).head(top_n_recommendations)

print("‚úÖ Local Recommendations:")
local_rec[['Destination Title', 'similarity_score']]
```

```python
def is_non_english(text):
    try:
        return langdetect.detect(text) != "en"
    except:
        return False
```

```python

def get_gemini_recommendation(query, hidden=False):
    """Calls Gemini API for travel advice."""
    prompt = f"User is looking for: {query} in Sri Lanka."
    if hidden:
        prompt += " Provide 5 'Hidden Gem' locations that are not very crowded."
    else:
        prompt += " Provide a list of the top 5 recommended destinations."
    
    try:
        response = gemini_model.generate_content(prompt)
        return response.text
    except Exception as e:
        return f"Gemini Error: {e}"

def hybrid_recommendation(query, filtered_df, all_embeddings):
    """
    Logic:
    1. If non-English -> Use Gemini
    2. Else -> Use SBERT
    3. If SBERT best score < threshold -> Use Gemini
    4. Always get Hidden Gems from Gemini
    """
    
    # Check language
    if is_non_english(query):
        return get_gemini_recommendation(query), get_gemini_recommendation(query, hidden=True)

    # SBERT Calculation
    query_emb = model.encode(query, convert_to_tensor=True)
    
    # IMPORTANT: We must calculate similarity against the WHOLE dataset 
    # and then filter, OR map the filtered indices correctly.
    all_similarities = util.cos_sim(query_emb, all_embeddings)[0].cpu().numpy()
    
    # Map similarities back to the filtered dataframe
    filtered_df = filtered_df.copy()
    # We use the index of the filtered_df to pick the right scores from the original list
    filtered_df['similarity_score'] = all_similarities[filtered_df.index]
    
    # Sort results
    local_results = filtered_df.sort_values(by='similarity_score', ascending=False).head(top_n_recommendations)
    
    max_score = local_results['similarity_score'].max() if not local_results.empty else 0

    # Decide if SBERT is enough
    if max_score < sbert_threshold:
        print(f"Low Confidence ({max_score:.2f}) -> Fetching from Gemini...")
        main_rec = get_gemini_recommendation(query)
    else:
        main_rec = local_results[['Destination Title', 'similarity_score']]

    # Always get Hidden Gems from Gemini
    hidden_gems = get_gemini_recommendation(query, hidden=True)
    
    return main_rec, hidden_gems

# --------------------------
# UI Logic
# --------------------------
query_box = widgets.Text(description='Query:', layout={'width': '80%'})
category_select = widgets.Dropdown(options=["All", "Family", "Couples", "Solo", "Business"], value='All', description='Category:')
recommend_btn = widgets.Button(description='Get Recommendations', button_style='success')
output = widgets.Output()

```

```python
def on_click(b):
    with output:
        clear_output()
        query = query_box.value.strip()
        if not query: return

        # Apply Category Filter
        filtered_docs = df.copy()
        if category_select.value != "All":
            cat_col = f"{category_select.value} %"
            if cat_col in filtered_docs.columns:
                filtered_docs = filtered_docs[filtered_docs[cat_col] > 0]

        # Get Hybrid Results
        main_rec, hidden_gems = hybrid_recommendation(query, filtered_docs, doc_embeddings)

        print("--- LOCAL DATABASE MATCHES (SBERT) ---")
        if isinstance(main_rec, pd.DataFrame):
            display(main_rec)
        else:
            print(main_rec)

        print("\n--- GEMINI HIDDEN GEMS ---")
        print(hidden_gems)

recommend_btn.on_click(on_click)
display(query_box, category_select, recommend_btn, output)
```
