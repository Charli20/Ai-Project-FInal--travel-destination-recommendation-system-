```python
import pandas as pd
import os
from langchain.docstore.document import Document
from langchain.text_splitter import CharacterTextSplitter
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain.vectorstores import Chroma

print("LangChain + Gemini OK ✅")

```

```python
df = pd.read_csv(r"C:\\NIBM\\Travel recommendation system\\Travel-Recommendation-system-in-Sri-Lanka-\\Data Preparation\\document_data\\documents.csv")
```

```python
import re

def text_preprocessing(text):
    text = text.lower()
    text = re.sub(r"<.*?>", " ", text)
    text = re.sub(r"[^\w\s]", " ", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

df["text_embedding"] = df["Destination Title"].astype(str) + ". " + df["Document Text"].astype(str)
df["clean_embedding"] = df["text_embedding"].apply(text_preprocessing)

print(df["clean_embedding"].iloc[0])
```

```python
documents = []

for _, row in df.iterrows():
    metadata = {
        "title": row["Destination Title"],
        "score": row["Destination Score"],
        "family": row["Family %"],
        "couples": row["Couples %"],
        "solo": row["Solo %"],
        "business": row["Business %"],
    }

    documents.append(
        Document(
            page_content=row["clean_embedding"],
            metadata=metadata
        )
    )
```

```python
embeddings = GoogleGenerativeAIEmbeddings(
    model="models/embedding-001"
)

db_travel = Chroma.from_documents(
    documents,
    embedding=embeddings,
    persist_directory="travel_chroma_db"
)
```

```python



from langchain.vectorstores import Chroma
import time

db_travel = Chroma(
    embedding_function=embeddings,
    persist_directory="travel_chroma_db"
)

BATCH_SIZE = 5  # VERY important for Gemini free tier

for i in range(0, len(documents), BATCH_SIZE):
    batch = documents[i:i + BATCH_SIZE]
    db_travel.add_documents(batch)
    time.sleep(15)  # wait to avoid rate limit

db_travel.persist()

```

```python
def get_travel_recommendations(query, top_k=10):
    results = db_travel.similarity_search(query, k=top_k)

    recommendations = []
    for doc in results:
        recommendations.append({
            "Destination": doc.metadata["title"],
            "Score": doc.metadata["score"]
        })

    return pd.DataFrame(recommendations)
```

```python
def get_hidden_gems(query, top_k=10, max_score=4.5):
    results = db_travel.similarity_search(query, k=30)

    gems = []
    for doc in results:
        if doc.metadata["score"] < max_score:
            gems.append({
                "Destination": doc.metadata["title"],
                "Score": doc.metadata["score"]
            })

        if len(gems) == top_k:
            break

    return pd.DataFrame(gems)
```

```python
query = "I need a beach place to relax"

print("✅ Recommended Destinations")
print(get_travel_recommendations(query))

print("\n🌟 Hidden Gems")
print(get_hidden_gems(query))
```
